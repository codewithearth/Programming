# Find_the_Runner-Up_Score!

```python
if __name__ == '__main__':
    n = int(input())
    arr = list(map(int, input().split()))
    maximum = max(arr)
    arr1 = []
    for i in arr:
        if i == maximum:
            continue
        else:
            arr1.append(i)
    print(max(arr1))

#2 3 6 6 5 
```
