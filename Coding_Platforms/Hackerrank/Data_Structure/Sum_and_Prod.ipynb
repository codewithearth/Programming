#**Sum and Prod**

```python
import numpy as np

#Size of Matrix
size_mtrx = list(map(int,input().split()))

#take values in list
A = []
for i in range(size_mtrx[0]):
    A.append(list(map(int,input().split())))

#Convert list into Array
a = np.array(A)

#Sum of Matrix
sum_mtrx = np.sum(a,axis = 0)

#Product of the Sum of Matrix
print(np.prod(sum_mtrx))
```
