#**Zeros and Ones**


```python
import numpy as np

list1 = list(map(int,input().split()))
list2 = []
list3 = []
list21 = []
list31 = []
list22 = []
list32 = []
if len(list1) == 3:
    for i in range(list1[0]*2):
        if i < list1[0]:
            list2.append(np.zeros((list1[1],list1[2]), dtype = np.int))
        
        else:
            list3.append(np.ones((list1[1],list1[2]), dtype = np.int))
        
    list2 = np.array(list2)
    print(list2) 
    list3 = np.array(list3)
    print(list3)
elif len(list1) == 4:
    for i in range(list1[0]*2):
        for j in range(list1[1]):
            if i == 0:
                list2.append(np.zeros((list1[2],list1[3]), dtype = np.int))
            elif i == list1[0]:
                list3.append(np.ones((list1[2],list1[3]), dtype = np.int))
        if i < list1[0]:
            list21.append(list2)
        else:
            list31.append(list3)    
    list21 = np.array(list21)
    print(list21)   
    list31 = np.array(list31)
    print(list31)
else:
    for i in range(2):
        if i == 0:
            print(np.zeros((list1[0],list1[1]), dtype = np.int))
        else:
            print(np.ones((list1[0],list1[1]), dtype = np.int))
```
