#**Mean, Var, and Std**

```python
import numpy as np

#Size of matrix
size_mtrx = list(map(int,input().split()))

#Values in matrix
A = []
for i in range(size_mtrx[0]):
    A.append(list(map(int,input().split())))

#Convert Matrix into Array
a = np.array(A)

#Mean of Matrix
print(np.mean(a,axis = 1))

#Variance of Matrix
print(np.var(a,axis = 0))

#Standard Deviation of Matrix
strd = np.std(a,axis = None,)
print(np.around(strd,decimals=11))
#1.11803398875
```
