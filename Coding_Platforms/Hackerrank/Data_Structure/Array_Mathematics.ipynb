#**Array Mathematics**

```python
import numpy as np

#Size of Matrix
size_mtrx = list(map(int,input().split()))

#First Matrix
A = []
for i in range(size_mtrx[0]):
    A.append(list(map(int,input().split())))
#Second Matrix
B = []
for j in range(size_mtrx[0]):
    B.append(list(map(int,input().split())))

#convert list to Array
a = np.array(A)
b = np.array(B)

#Add a and b
print(np.add(a,b))

#Subtract a and b
print(np.subtract(a,b))

#Multiply a and b
print(np.multiply(a,b))

#Divide a and b
Div1 = np.divide(a,b)
print(np.around(Div1.astype(int)))

#Mod of a and b
print(np.mod(a,b))

#Power of a and b
print(np.power(a,b))
```
