# Nested Lists


```python
if __name__ == '__main__':
    l = []
    l1 = []
    l2 = []
    m = []
    m1 = []
    n1 = []
    n2 = []
    q = []
    for i in range(int(input())):
        name = input()
        l1.append(name)
        score = float(input())
        l2.append(score)
        l = [l1[i],l2[i]]
        m .append(l)   
    sorted(m)
    for n in range(len(m)):
        m1.append(m[n][1])
    minimum = min(m1)
    for m2 in m1:
        if m2 == minimum:
            continue
        n1.append(m2)
    res = min(n1)
    for o in range(len(m)):
        if res == m[o][1]:
            n2.append(m[o][:])
        else:
            continue
    for p in range(len(n2)):
        q.append(n2[p][0])
    q1 = sorted(q)
    for r in q1:
        print(r)

```
