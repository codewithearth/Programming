#**Min and Max**

```python
import numpy as np

#Size of Matrix
size_mtrx = list(map(int,input().split()))

#Input value
A = []
for i in range(size_mtrx[0]):
    A.append(list(map(int,input().split())))

#Convert Matrix into array
a = np.array(A)

#Minimum in Matrix along ROW
min = np.min(a, axis = 1)

#Maximum in the Minimum of Matrix
print(np.max(min))
```
