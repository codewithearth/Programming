#**Problem Statement**

You're given a number N. Print the first N lines of the below-given pattern.
```
    *
   **
  ***
 ****
*****
```

##**Input:**
* First-line will contain the number N.

##**Output:**
Print the first N lines of the given pattern.

##**Constraints**

* 1≤N≤200

##**Sample Input 1:**
```
4
```

##**Sample Output 1:**
```
   *
  **
 ***
****
```

##**Sample Input 2:**
```
2
```

##**Sample Output 2:**
```
 *
**
```

##**EXPLANATION:**
* In the first example, we'll print the first 4 lines of the given pattern.
* In the second example, we'll print the first 2 lines of the given pattern


#**Code:**

````python
# create rev_star recursive function
def rev_star(n,m):

    # if n is -1 then return the function
    if n == -1:
        return

    # print space before star
    # print star pattern
    print(" "*n,end="")
    print("*"*m)

    # call the recursive function
    rev_star(n-1,m+1)

# input number
a = int(input())

# set b is 1
b = 1

# call rev_star function and put a and b value
rev_star(a-1,b)
````
